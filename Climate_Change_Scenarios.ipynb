###Concepts covered:
Climate dynamics model: The input (or forcing) to the climate dynamics model is yearly atmospheric CO 2 concentrations,
measured in ppm. The output of the climate dynamics model is the yearly average temperature
increase over pre-industrial temperatures in °C
-The first part computes how much extra energy is warming the atmosphere due to climate change and what the long-term temperature effect of
- that extra energy would be. The second part computes the predicted yearly global average temperature increase over time.
-  The amount of extra energy caused by rising CO 2 concentrations is called the radiative forcing andnis measured in W/m 2 .
- The equation to compute this variable is rft Co2= 5.35 ln(Ct/Cpre)
- The total effect of global warming, that is all greenhouse gases, is computed as rft= rft co2 + rft other ggs.
   - lambda is called the climate sensitivity
   - We will use a very simple delay formulation. For each time step we will first compute the difference between the temperature
   in the previous year and the temperature that the system would reach if radiative forcing were to be held constant for a very long time.
   We then assume that the actual temperature will warm by a very small fraction of this computed difference.


Parameters:
- CO2 concentration (pre-industrial) (ppm) = 275
- lambda (climate sensitivity) = 0.8
- mu (delay of warming)=1/66

```python
from google.colab import files
import pandas as pd

# Assuming the uploaded file is named 'CCE Lab 01 - Forcings.xlsx'
file_name = 'CCE Lab 01 - Forcings.xlsx'
df = pd.read_excel(file_name)
#display(df)
#columns are numbers from 0 to 287
df.columns=range(len(df.columns))
display(df)

CO2conc=df.iloc[6,:]
Radforcings=df.iloc[7,:]
year=df.iloc[2,:]
year
```

```python
GivenForcings=pd.DataFrame({'Year':year, 'CO2 concentrations (ppm)':CO2conc,'Radiative forcing other gases (W/m^2)':Radforcings})
GivenForcings=GivenForcings.transpose()

# Clean up the column names (remove the first column which contains the original row labels)
GivenForcings=GivenForcings.iloc[:, 1:]
display(GivenForcings)
```

```python
import math as math
import pandas as pd

#parameters
preindco2=275
lambda1=0.8
mu=1/66

#Variables
# Calculate the natural logarithm for each value in the Series
log_co2_conc = GivenForcings.loc['CO2 concentrations (ppm)'].apply(lambda x: math.log(x / 275))
radforcing_co2 = log_co2_conc *5.35
total_gghg=radforcing_co2 + GivenForcings.loc['Radiative forcing other gases (W/m^2)']
Eqwarming=lambda1* total_gghg
#tempincrease1=0.9+mu*(Eqwarming-0.9)
tempincrease = [0.9,0.9]
for x in range(1, 286):
  next_value = tempincrease[x - 1] + mu * (Eqwarming[x] - tempincrease[x - 1])
  tempincrease.append(next_value)
temp_over=pd.Series(tempincrease)


Temp_over_preind=pd.DataFrame({'Year':year, 'CO2 concentrations (ppm)':CO2conc,
                                'Radiative forcing other gases (W/m^2)':Radforcings,
                                'Total Radiative Forcing from CO2': radforcing_co2,
                                'Effects from gghg': total_gghg,
                                'Equilibrium Warming': Eqwarming,
                                'Temperature Increase over pre-industrial': temp_over})
Temp_over_preind.iloc[1:,:]
# .iloc[1:,:] drops first row
#.iloc[:, 1:] cuts first column
```
